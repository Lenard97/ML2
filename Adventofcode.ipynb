```python
import pandas as pd

# Pfad zur Excel-Datei
excel_path = r'C:\Users\g065308\Drive Merkur.works\privat\adventofcode23\01012023.xlsx'

# Initialisiere die Summe
gesamtsumme = 0

try:
    # Lese die Excel-Datei
    df = pd.read_excel(excel_path, header=None, names=['Column1'])

    # Iteriere über jede Zeile im DataFrame
    for index, row in df.iterrows():
        input_string = str(row['Column1'])  # Verwende den tatsächlichen Spaltennamen

        # Extrahiere nur Ziffern aus dem aktuellen Input
        ziffern = [char for char in input_string if char.isdigit()]

        # Wenn mehr als zwei Ziffern vorhanden sind, gib die erste und letzte aus
        if len(ziffern) > 2:
            erste_ziffer = ziffern[0]
            letzte_ziffer = ziffern[-1]
            print(f"{erste_ziffer}{letzte_ziffer}")

            # Füge die Ziffern zur Summe hinzu
            gesamtsumme += int(erste_ziffer + letzte_ziffer)
        elif len(ziffern) == 2:
            print(f"{''.join(ziffern)}")

            # Füge die Ziffern zur Summe hinzu
            gesamtsumme += int(''.join(ziffern))
        elif len(ziffern) == 1:
            print(f"{''.join(ziffern)}{''.join(ziffern)}")

            # Füge die Ziffern zur Summe hinzu
            gesamtsumme += int(''.join(ziffern) * 2)

except Exception as e:
    print(f"Fehler beim Lesen der Excel-Datei: {e}")

# Gib die Gesamtsumme aus
print(f"Gesamtsumme: {gesamtsumme}")
```
